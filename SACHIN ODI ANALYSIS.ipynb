```python
Opposition_vs_100 = pd.pivot_table(data=sachin,index='Opposition',columns='Hundred',values='Runs',aggfunc='count',fill_value=0)[1]
Opposition_vs_100
# creating a pivot table that shows the count of centuries (100 runs or more) scored against each opposition team.
# fill_value=0 sets the fill value for missing entries to 0.
# [1] is used to select only the column corresponding to centuries.
```

```python
Position_wise_runs = sachin.groupby('Pos')['Runs'].sum().sort_values(ascending=False)
Position_wise_runs
# grouping the data by the 'Pos' (Position) column, calculates the sum of 'Runs' for each position, and sorts the result in descending order.
```

```python
Innings_wise_runs = sachin.groupby('Inns')['Runs'].sum().sort_values(ascending=False)
Innings_wise_runs
# grouping the data by the 'Inns' (Innings) column, calculates the sum of 'Runs' for each innings, and sorts the result in descending order.
```

```python
total_runs_by_opp = pd.crosstab(index=sachin['Opposition'],columns='Total Runs',values=sachin['Runs'],aggfunc='sum').sort_values(by='Total Runs',ascending=False)
total_runs_by_opp = total_runs_by_opp['Total Runs']
total_runs_by_opp
# createing a cross-tabulation that shows the total runs scored against each opposition team.
```

```python
import matplotlib.pyplot as plt # For plotting
import seaborn as sb # For data Visualization
```

```python
plt.hist(sachin['Runs'],bins=10)
plt.title('Distribution of Runs')
plt.xlabel('Runs')
plt.ylabel('Frequency')
plt.show()
# The histogram shows that the most common run score is between 25 and 50 runs. There are also a significant number of scores between 50 and 75 runs, and between 75 and 100 runs. Fewer scores are above 100 runs
```

```python
Total_runs_yearly = pd.crosstab(index=sachin['Year'],columns='Total Runs',values=sachin['Runs'],aggfunc='sum').sort_values(by='Total Runs',ascending=False)
Total_runs_yearly= Total_runs_yearly['Total Runs']
Total_runs_yearly
# Total_runs_yearly shows the total runs scored in each year, with the years sorted in descending order based on the total runs.
```

```python
Total_runs_yearly.plot(kind='barh',color='green')
plt.title('Total Runs Per Year')
plt.xlabel('Total Runs')
plt.ylabel('Year')
plt.show()
# The horizontal bar graph shows that Sachin scored the highest runs by year in 2007. 
```

```python
sr = round(pd.crosstab(index=sachin['Opposition'],columns='Strike Rate',values=sachin['SR'],aggfunc='mean').sort_values(by='Strike Rate',ascending=False),2)
sr = sr['Strike Rate']
sr
# sr showing the average strike rate against different opposition teams.
```

```python
sr.plot(kind='barh')
plt.title('Opposition wise Average Strike Rate')
plt.xlabel('Strike Rate')
plt.ylabel('Opposition')
plt.show()
# Sachin has the highest average strike rate against Netherlands and least aganist West Indies.
```

```python
plt.figure(figsize=(8,5))
sb.regplot(data=sachin,x='Runs',y='Balls Faced',scatter=True)
plt.title('Relation between Runs Vs Balls Faced')
plt.xlabel('Runs')
plt.ylabel('Balls Faced')
plt.show()
# regplot gives us a linear regression line
# the graph shows that there is a positive correlation between runs scored and balls faced in cricket.
# However, the relationship is non-linear, with batsmen scoring runs at a higher rate towards the end of an innings.
# There are a few possible explanations for the non-linear relationship between runs scored and balls faced. One possibility is that batsmen are more likely to take risks early in their innings in order to score quickly. Another possibility is that bowlers become more tired towards the end of an innings, which makes it easier for batsmen to score runs.
```

```python
sachin[['Runs','Balls Faced']].corr()
# he correlation coefficient is approximately 0.95, which indicates a strong positive correlation between the number of runs scored and the number of balls faced. This suggests that as the number of runs increases, the number of balls faced also tends to increase, and vice versa.
```

```python
plt.figure(figsize=(8,5))
sb.regplot(data=sachin,x='Mins',y='Runs',color='green')
plt.title('Relation between Runs scored by Mins')
plt.xlabel('Mins')
plt.ylabel('Runs')
plt.show()
plt.show()
# The graph shows a positive correlation between runs scored and minutes played. This means that the longer a batsman plays, the more runs they are likely to score. This is because batsmen have more time to score runs if they play for more minutes.
```

```python
sachin_corr = sachin[['Runs','Balls Faced','4s','6s','SR']].corr()
sachin_corr
# Shows the correlation matrix for the variables 'Runs', 'Balls Faced', '4s', '6s', and 'SR'. This matrix shows the pairwise correlations between these variables.
```

```python
plt.figure(figsize=(8,5))
sb.heatmap(sachin_corr,annot=True,fmt='.2f')
plt.title('Correlation between Runs, Balls Faced, 4s, 6s and SR')
plt.show()
# The correlation matrix shows that runs are strongly correlated with balls faced (0.95), 4s (0.91), and 6s (0.59). This means that batsmen who face more balls and hit more 4s and 6s are more likely to score more runs.
# SR is also strongly correlated with 4s (0.83) and 6s (0.61). This means that batsmen who hit more 4s and 6s are more likely to have a higher strike rate.
# The correlation matrix shows that runs, balls faced, 4s, 6s, and SR are all highly correlated.
# The correlation matrix can be used to help coaches and captains identify batsmen who are in good form and to develop strategies for batsmen to improve their performance.
```
